```python
import cv2

!wget -q -O haarface.xml https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade = cv2.CascadeClassifier('haarface.xml')
print("Haar cascade loaded:", face_cascade.empty() == False)
```

```python
from google.colab import files
uploaded = files.upload()

import cv2
import numpy as np

image_path = list(uploaded.keys())[0]
print("Uploaded image:", image_path)
```

**Example Dataset**

```python
from IPython.display import display, Image as IPImage
import cv2

def detect_and_display(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Haar detection
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4)

    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)

    _, buf = cv2.imencode('.jpg', img)
    display(IPImage(data=buf.tobytes()))

    print(f"Total faces detected: {len(faces)}")

detect_and_display(image_path)
```
